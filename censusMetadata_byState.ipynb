```python
def rename_cols(df):
    cols = df.columns
    indices = sorted([[i, name, states[name]] for i,name in enumerate(cols) if name in states.keys()])
    
    newcols = []
    for i in range(0, len(indices)):
        ### first, get the starting index + the state and abbreviation
        s_ind, st, abbv = indices[i][0], indices[i][1], indices[i][2]
        if (i+1) == len(indices):
            e_ind = df.shape[1]
        else:
            e_ind = indices[i+1][0]
        
        print(f'{s_ind}, {e_ind}')
        st_cols = cols[s_ind:e_ind]
        newcol_state = [abbv + "_" + col.replace(u'\xa0', u'') for col in st_cols]
        newcols.extend(newcol_state)
    df.columns = newcols
    return df
        
```
